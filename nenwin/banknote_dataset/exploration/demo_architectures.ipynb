## Demonstation of architectures for the training.
Helps with testing if they are created as intended.

```python
from nenwin.banknote_dataset.architectures import ARCHITECTURES, gen_architecture
from nenwin.plot_model import plot_model
```

```python
model_a, _, _ = gen_architecture(ARCHITECTURES.A)
fig, ax = plot_model(model_a)
ax.set_title("Model A", weight="bold");
```

```python
model_b, _, _ = gen_architecture(ARCHITECTURES.B)
fig, ax = plot_model(model_b)
ax.set_title("Model B", weight="bold")
ax.set_xlim([-11, 11])
ax.set_ylim([-11, 11]);
```

```python
model_c, _, _ = gen_architecture(ARCHITECTURES.C)
fig, ax = plot_model(model_c)
ax.set_title("Model C", weight="bold")
ax.set_xlim([0, 6])
ax.set_ylim([0, 5]);
```
